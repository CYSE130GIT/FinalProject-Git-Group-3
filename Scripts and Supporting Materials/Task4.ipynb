```python
import subprocess
import schedule
from datetime import datetime
import time
#Function to Run nmap
def run_nmap(target):
    #Timestamp for the outputed scan file
    timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
    #Output filenameing scheme
    filename = f'nmap_scan_{timestamp}.txt'
    #Running nmap with the arguments -sV, -v, and --version intensity 5
    #Version intensity 5 is to let the process actually run on my loopback address on windows
    #Not needed on linux machines
    result = subprocess.run(['nmap', '-sV', '--version-intensity', '5', '-v', target], capture_output=True, text=True)
    #write the result of the scan into the file and save/print it out in the console
    with open(filename, 'w') as file:
        file.write(result.stdout)
    print(result.stdout)

#Schedules the nmap scan to run every hour
schedule.every(1).hours.do(run_nmap, '127.0.0.1')  # Scan localhost
#Runs the first nmap scan
run_nmap('127.0.0.1')
#Keeps it running
while True:
    schedule.run_pending()
    time.sleep(1)
```

```python
from scapy.all import *
#Function to print results of the sniff if they are TCP/IP Packets
def monitor_packets(pkt):
    if pkt.haslayer(TCP) and pkt.haslayer(IP):
        print(f"Source IP: {pkt[IP].src}, Destination IP: {pkt[IP].dst}")

capture=sniff(prn=monitor_packets, count=10)  # Capture 10 packets
#Save capture to pcap file
wrpcap('Test.pcap', capture)

```
